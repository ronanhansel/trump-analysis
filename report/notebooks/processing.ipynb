```python
import pandas as pd
```

```python
df = pd.read_csv('../data/truth_social_posts_final_2025.csv')
df.drop(columns=['content_html', 'date'], inplace=True)
```

```python
df.head()
```

```python
# Create boolean columns for repub and media categories
df['is_retweet'] = df['content_text'].fillna('').str.startswith('RT: ')

df['is_media_only'] = (
    (df['deleted_flag'] == True) |
    (df['content_text'].isna()) |
    (df['content_text'] == '') |
    (df['content_text'].str.strip() == '')
)
```

```python
import re

def clean_content_text(text):
    if pd.isna(text):
        return text
    # Remove 'RT: ' at the start followed by a URL and optional space
    text = re.sub(r'^RT:\s*(https?://\S+)\s*', '', text)
    # Remove all URLs anywhere in the text
    text = re.sub(r'https?://\S+', '', text)
    # Remove extra spaces left by URL removal
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df['content_text'] = df['content_text'].apply(clean_content_text)

df = df[df['deleted_flag'] == False]

df.to_csv('../data/truth_social_posts_final_2025_cleaned.csv', index=False)
```
