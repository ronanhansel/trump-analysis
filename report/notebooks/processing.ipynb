```python
import pandas as pd
```

```python
df = pd.read_csv('../data/truth_social_posts_final_2025.csv')
df.drop(columns=['content_html', 'date'], inplace=True)
```

```python
df.head()
```

```python
# Create boolean columns for repub and media categories
df['is_retweet'] = df['content_text'].fillna('').str.startswith('RT: ')

df['is_media_only'] = (
    (df['deleted_flag'] == True) |
    (df['content_text'].isna()) |
    (df['content_text'] == '') |
    (df['content_text'].str.strip() == '')
)
```

```python
import re

def remove_rt_link(text):
    if pd.isna(text):
        return text
    # Match 'RT: ' at the start, followed by a URL, then a space or end, then the rest
    match = re.match(r'^RT:\s*(https?://\S+)\s*(.*)', text)
    if match:
        return match.group(2).lstrip()
    return text

df['content_text'] = df['content_text'].apply(remove_rt_link)

df = df[df['deleted_flag'] == False]

df.to_csv('truth_social_posts_final_2025_cleaned.csv', index=False)
```
